# Plot of accuracy versus prominence threshold

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

import seaborn as sns
sns.set_context('paper', font_scale=2)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../experiments/01_accuracy_vs_prom/accuracy_vs_Prom.csv', index_col=0)
```

```python
df
```

```python
plt.plot(df.n_lines, df.std_dev, 'o-')
plt.xlabel('Number of lines')
plt.ylabel('Per-pixel error')
#plt.xscale('log')
plt.xlim(0)
```

```python
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

```python
plt.plot(df.prominences, df.std_dev*100, 'o-', color='#2980b9', markersize=8)
plt.axhline(1.2, color='#e67e22', linestyle='dashed', label='Systematic error floor')
plt.xlabel('Prominence threshold $(P_{rom})$', fontsize=16)
plt.ylabel('Error $(\% \; \mathrm{pixel}^{-1})$', fontsize=16)
plt.xlim(0.2, 0.0005)
plt.ylim(0, 5)
plt.xscale('log')
plt.legend(fontsize=13)
plt.title('PHOENIX Native-res HPF clone');
plt.savefig('../experiments/01_accuracy_vs_prom/accuracy_vs_prom.png', 
            dpi=300, bbox_inches='tight', transparent=False)
```
